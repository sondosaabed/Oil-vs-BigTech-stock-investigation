## In class group activity

```python
import numpy as np 
import pandas as pd 
```

```python
df = pd.read_csv('/kaggle/input/data-science-salaries/ds_salaries.csv')
```

```python
df.head()
```

1. Display the top 10 countries with the highest salaries.
- Group by country (get salary)
- datafram from (groupby median -> sort values -> iloc 10

```python
grouped_by_location = df['salary_in_usd'].groupby(df['company_location'])
pd.DataFrame(grouped_by_location.median()).sort_values(by='salary_in_usd', ascending= False).iloc[:10]
```

2. For each job title, present the top 5 countries with the higfhest salaries.

```python
grouped_by_jobTitle = df[['job_title','salary_in_usd','company_location']].groupby(by=['job_title', 'company_location']).agg('max')
display(grouped_by_jobTitle)
```

3. Determmine the country that offers the highest salary for each job title compared to other countries.

```python
highest_salary_by_job_title = df.loc[df.groupby('job_title')['salary_in_usd'].idxmax()]
display(highest_salary_by_job_title[['job_title','company_location','salary_in_usd']])
```
